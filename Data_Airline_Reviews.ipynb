# Cleaning the Dataset

### In our dataset there are many rows in which all the values of  the coloums are missing or NaN,  and also we are having some duplicate reviews . So let's first remove all those rows and then we will perform the analysis.

```python
# checking the rows for the NaN values

NaN_rows = airline_data.isnull().any(axis = 1)
print(NaN_rows)
```

```python
# checking for the duplicate reviews or data 

airline_data.duplicated()
```

### So as we are having duplicated reviews also then now we go for the getting the number of the duplicated reviews including NaN.

```python
duplicate = airline_data[airline_data.duplicated()]
duplicate
```

```python
# Removing all the NaN reviews (excluding the repeated reviews)

without_NaN = airline_data.dropna()
without_NaN
```

```python
# Removing all the duplicate or repeated reviews

# as we have already removed or dropped the NaN values, now for removing the duplicate data we will work on the variable
# "without_NaN" for making our dataset cleaned or more efficient for analysis.

efficient_data = without_NaN.drop_duplicates()
efficient_data
```

## Now we got the dataset cleaned and more efficient for performing analysis.

# Analysing the Review of the customers

### For further analysis we will be using "efficient_data" as our dataset.

## Let's first see the total number of reviews for each airline.

```python
efficient_data.airline.value_counts()
```

### Now as we are having total 81 different airlines, so let's print the bar graph for the first 5 and last 5 

```python
import matplotlib.pyplot as plt
```

```python
efficient_data.airline.value_counts().head().plot.bar(color = ['b','y','g','m','c'])

plt.xlabel("Name of Airways")
plt.ylabel("Number of Reviews")

plt.title("First Five")
```

## From the bargraph above we conclude that British Airways has the largest number of reviews provided by users.

```python
efficient_data.airline.value_counts().tail().plot.bar(color = ['k','r','y','b','g'])

plt.xlabel("Name of Airways")
plt.ylabel("Number of Reviews")

plt.title("Last Five")
```

## Again from the bargraph above we conclude that Germanwings, QantasLink,airBaltic has the smallest and equal number of reviews provided by the users.
