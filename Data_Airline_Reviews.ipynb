# Cleaning the Dataset

### In our dataset there are many rows in which all the values of  the coloums are missing or NaN,  and also we are having some duplicate reviews . So let's first remove all those rows and then we will perform the analysis.

```python
# checking the rows for the NaN values

NaN_rows = airline_data.isnull().any(axis = 1)
print(NaN_rows)
```

```python
# checking for the duplicate reviews or data 

airline_data.duplicated()
```

### So as we are having duplicated reviews also then now we go for the getting the number of the duplicated reviews including NaN.

```python
duplicate = airline_data[airline_data.duplicated()]
duplicate
```

```python
# Removing all the NaN reviews (excluding the repeated reviews)

without_NaN = airline_data.dropna()
without_NaN
```

```python
# Removing all the duplicate or repeated reviewsairline

# as we have already removed or dropped the NaN values, now for removing the duplicate data we will work on the variable
# "without_NaN" for making our dataset cleaned or more efficient for analysis.

efficient_data = without_NaN.drop_duplicates()
efficient_data
```

```python
efficient_data.isnull().sum()
```

```python
duplicate = efficient_data[airline_data.duplicated()]
duplicate
```

## Now we got the dataset cleaned and more efficient for performing analysis.

```python
efficient_data["review_date"]=pd.to_datetime(efficient_data["review_date"])
```
